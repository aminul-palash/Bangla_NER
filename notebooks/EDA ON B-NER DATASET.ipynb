```python
dataset = []
def compute_new_column(row):
    # Example: Return the double of the Age column value
    data = ast.literal_eval(row['word_tag'])
    
    for item in data:
        dataset.append(item)
        
    dataset.append(("।",'O'))
    dataset.append("\n")
    return True
```

```python
df['NewColumn'] = df.apply(compute_new_column, axis=1)
```

```python
dataset
```

```python
with open('palash_output.txt', 'w') as file:
    
    for line in dataset:
        try:
            file.write(str(line[0]) + "\t" + str(line[1] + "\n"))
        except:
            file.write(str(line) )
```
